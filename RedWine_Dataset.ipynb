```python
## visualization 
df.quality.value_counts()
```

CONCLUSION : It is an imbalanced Dataset

```python
# Conclusion - It is an imbalanced dataset
df.quality.value_counts().plot(kind='bar')
plt.xlabel('Wine Quality')
plt.ylabel('Count')
plt.show()
```
